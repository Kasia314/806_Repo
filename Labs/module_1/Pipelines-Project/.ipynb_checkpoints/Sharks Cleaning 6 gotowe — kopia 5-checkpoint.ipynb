```python
def clean_colnames(df):
    '''Clean columns names when passing a pandas dataframe: params (df - dataframe)'''
    col_clean = []
    for col in df.columns:
        col = col.strip().upper()
        col_clean.append(col)
        
    df.columns = col_clean
    return df.columns

clean_colnames(data1)
```
