```python
mylist = ["leg","hand"]
pattern = '|'.join(mylist)
import re
def pattern_searcher(search_str:str, search_list:str):

    search_obj = re.search(search_list, search_str)
    if search_obj :
        return_str = search_str[search_obj.start(): search_obj.end()]
    else:
        return_str = 'NA'
    return return_str

data3['matched_str'] = data3['Injury'].apply(lambda x: pattern_searcher(search_str=x, search_list=pattern))
```
