#### The data we will use in this challenge is:

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x, y) and (x, z).
There are 2 ways of doing this. Do in both ways.

**Hint**: Check out the nrows, ncols and index arguments of subplots. 

Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# Method 1
plt.plot(x,y, label='linear')
plt.plot(x,z, label='quadric')

######### ozdobniki:
plt.xlabel('x label')
plt.ylabel('y label')

plt.title('Linear vs Quadratic')
plt.legend()

plt.grid()

plt.show()
```

#### Use plt.subplots(nrows=1, ncols=2) to create the plot.

```python
# Method 2
fig, axes=plt.subplots(1,2)

axes[0].plot(x,y, label='linear')
axes[1].plot(x,z, label='quadric')

############### ozdobniki:
axes[0].set_xlabel('x label')
axes[0].set_ylabel('y label')

axes[1].set_xlabel('x label')
axes[1].set_ylabel('y label')

axes[0].set_title('Linear one')
axes[1].set_title('Quadratic one')

plt.legend()

plt.show()
```

#### Use your previous code but now, resize your plot.
**Hint**: Add the figsize argument in plt.subplots().

If you want, try to add a title to the plot or even axes labels. You can also play with the fontweight and fontsize of the titles and labels. 

```python
# your code here

import pylab as plb
import matplotlib.pyplot as mpl

######### size:
plt.figure(figsize=(10,10))
######### plot:
plt.plot(x,y, label='linear')
plt.plot(x,z, label='quadric')

######### ozdobniki:
csfont = {'fontname':'Comic Sans MS'}
hfont = {'fontname':'Comic Sans MS'}
plb.rcParams['font.size'] = 14
mpl.rcParams['font.size'] = 12


plt.xlabel('x label',**hfont)
plt.ylabel('y label',**hfont)


plt.title('Linear vs Quadratic',**csfont)


plt.legend()
plt.grid()
plt.show()
```

#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here
k= np.exp(x)

plt.plot(x,z, label='quadric')
plt.plot(x,k, label='exponential')
```

```python
fig, axes=plt.subplots(1,2)

axes[0].plot(x,z, label='quadric')
axes[0].plot(x,k, label='exponential')

axes[1].plot(x,z, label='quadric')
axes[1].plot(x,k, label='exponential')

############# set scales:
#set_xscale
#set_yscale


############### ozdobniki:
axes[0].set_xlabel('x label')
axes[0].set_ylabel('y label')
plt.legend()

axes[1].set_xlabel('x label')
axes[1].set_ylabel('exp(x)')
plt.xscale('log')
plt.yscale('log')
plt.legend()

axes[0].set_title('normal scale')
axes[1].set_title('log scale')



plt.show()
```
